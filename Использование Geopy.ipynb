```python
import pandas as pd
import networkx as nx
import folium
from geopy.distance import geodesic
from folium import plugins
```

**1. Выбрать (скачать) самостоятельно датасет (например, с kaggle.com) и построить граф с географической привязкой данных**

```python
df = pd.read_csv("C:\\Users\\Yo\\Downloads\\worldcitiespop.csv")
df = df[df['Country'] == 'ad'].head(100)
df.head()
```

**2. Для решения задания использовать методы библиотеки networkx (источник: networkx.org) или иные библиотеки**

```python
G = nx.Graph()

for _, row in df.iterrows():
    city = row['City']
    lat = row['Latitude']
    lng = row['Longitude']
    G.add_node(city, pos=(lat, lng))

print(list(G.nodes))
```

```python
for i, row1 in df.iterrows():
    city1 = row1['City']
    lat1 = row1['Latitude']
    lng1 = row1['Longitude']
    
    for j, row2 in df.iterrows():
        
        city2 = row2['City']
        
        if city1 != city2:
            lat2 = row2['Latitude']
            lng2 = row2['Longitude']
            distance = geodesic((lat1, lng1), (lat2, lng2)).kilometers
            G.add_edge(city1, city2, weight=distance)

print(f'Количество рёбер в графе: {len(G.edges)}')
```

**3. Рассчитать среднюю степень вершины, средний минимальный путь и диаметр полученного графа**

```python
avg_degree = sum(dict(G.degree()).values()) / len(G.nodes)
print(f'Средняя степень вершины: {avg_degree}')
```

```python
avg_shortest_path = nx.average_shortest_path_length(G)
print(f'Средний минимальный путь: {avg_shortest_path}')
```

```python
diameter = nx.diameter(G)
print(f'Диаметр графа: {diameter}')
```

**4. Проанализировать и выявить закономерности при их наличии (например, специфика регионов и т.п.)**

```python
center_lat, center_lng = df['Latitude'].mean(), df['Longitude'].mean()
m = folium.Map(location=[center_lat, center_lng], zoom_start=12)
```

```python
for node, (lat, lng) in nx.get_node_attributes(G, 'pos').items():
    folium.CircleMarker(location=[lat, lng], radius=5, color='red', fill=True, fill_color='red').add_to(m)
```

```python
for edge in G.edges:
    city1, city2 = edge
    lat1, lng1 = G.nodes[city1]['pos']
    lat2, lng2 = G.nodes[city2]['pos']
    folium.PolyLine([(lat1, lng1), (lat2, lng2)], color='blue', weight=1).add_to(m)
```

```python
plugins.MarkerCluster().add_to(m)
```
